```python
def make_row(lst):
  out = []
  out.append('<tr>')
  for item in lst:
    out.append('<td>' + item + '</td>')
  out.append('</tr>')
  return '\n'.join(out)
```

```python
from random import choice, sample, shuffle

yes_examples = []
no_examples = []
height = 'tall medium-height short'.split()
build = 'slim heavy average-weight'.split()
hair = 'bald short straight wavy curly coily'.split()
age = 'young middle-aged old'.split()
sex = 'male female'.split()

for h in height:
  for b in build:
    for c in hair:
      for a in age:
        for s in sex:
          out = [h, b, c, a, s]
          if 'tall' in out and 'wavy' in out and 'male' in out:
            decision = 'YES'
          else:
            decision = choice(['NO', 'NO', 'NO', 'YES'])
          vector = out + [decision]
          if vector[-1] == 'YES':
            yes_examples.append(vector)
          else:
            no_examples.append(vector)
```

```python
len(no_examples), len(yes_examples)
```

```python
yes_out = sample(yes_examples, 35)
no_out = sample(no_examples, 65)
data = yes_out + no_out
shuffle(data)
for d in data:
  print(d)
```
